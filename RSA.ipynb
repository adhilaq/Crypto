```python
import random

def is_prime(num):
    if num == 2:
        return True
    if num < 2 or num % 2 == 0:
        return False
    for n in range(3, int(num**0.5)+2, 2):
        if num % n == 0:
            return False
    return True

def generateRandomPrime():
    while(1):
        random_prime = random.randint(0,10000)
        if is_prime(random_prime):
            return random_prime
        
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)
    
def generate_key():
    p = generateRandomPrime()
    q = generateRandomPrime()
    
    n = p*q
    print("n ",n)
    phi = (p-1) * (q-1) 
   
    e = random.randint(1, phi)
    g = gcd(e,phi)
    while g != 1:
        e = random.randint(1, phi)
        g = gcd(e, phi)
        
    print("e=",e," ","phi=",phi)
  
    d = egcd(e, phi)[1]
    
    d = d % phi
    if(d < 0):
        d += phi
    return ((e,n),(d,n))

def decrypt(cipher_text,private_key):
    key,n = private_key
    text = []
    for char in cipher_text:
        text.append(chr(pow(char,key,n)))
    return "".join(text)

def encrypt(text,public_key):
    key,n = public_key
    cipher_text = []
    for char in text:
        cipher_text.append(pow(ord(char),key,n))
    return cipher_text


plain_text = input("Enter plaintext:")
public_key,private_key = generate_key() 
print("Public:", public_key)
print("Private:", private_key)


cipher_text = encrypt(plain_text, public_key)
print("Encrypted text:",cipher_text)
plaintext = decrypt(cipher_text, private_key)
print("Decrypted text=",plaintext)
```
